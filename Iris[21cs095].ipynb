```python
X = data.drop(['Id', 'Species'], axis=1)
y = data['Species']
```

```python
print(X.shape)
```

```python
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
score = logreg.score(X_test, y_test)
print(score)
```
